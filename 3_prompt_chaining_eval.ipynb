Create an outline, blog using a title as input and then evaluate the blog with an integer score

# start - outline - blog - evaluate - end

start - generate_outline - generate_blog - evaluate using integer scor - end

```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv
from typing import TypedDict

import os
load_dotenv()
api_key = os.environ.get("OPENAI_API_KEY")
```

```python
model = ChatOpenAI()
```

```python
#create a state
class BlogState(TypedDict):
    topic: str
    outline: str
    blog: str
    eval: int
```

```python
def create_outline(state: BlogState) -> BlogState:
    #generate the outline for the blog using the topic from the state
    topic = state['topic']
    prompt = f"create a detailed outline for the blog on the topic: {topic}"
    outline = model.invoke(prompt).content
    state['outline'] = outline    
    return state
```

```python
def create_blog(state: BlogState) -> BlogState:
    #generate the blog using the outline from the state
    title = state['topic']
    outline = state['outline']
    prompt = f"Create a detailed blog based on the following title: {title} using the following outline: \n {outline}"
    blog = model.invoke(prompt).content
    state['blog'] = blog
    return state



```

```python
def evaluate_blog(state: BlogState) -> BlogState:
    #evaluate the blog using the title and generate a score between 1-10
    title = state['topic']
    outline = state['outline']
    blog = state['blog']
    prompt = f"Evaluate the blog using the following title: {title} and outline: {outline} and blog: {blog}. Generate a score between 1-10"
    eval_score = model.invoke(prompt).content
    state['eval'] = eval_score
    return state

```

```python
#create a graph
graph = StateGraph(BlogState)

# create nodes

graph.add_node("create_outline", create_outline)
graph.add_node("create_blog", create_blog)
graph.add_node("evaluate_blog", evaluate_blog)

# create edges
graph.add_edge(START, "create_outline")
graph.add_edge("create_outline", "create_blog")
graph.add_edge("create_blog", "evaluate_blog")
graph.add_edge("evaluate_blog", END)

workflow = graph.compile()
initial_state = {"topic": "Agentic AI using Langgraph and Langchain handson project"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
